# <mark>NTA</mark>

```python
nta_dict = dict(zip(nta['ICD-10-CM Code'], nta['Comorbidity Description']))
```

```python
nta_dict = {}
for index, row in nta.iterrows():
    key = row['ICD-10-CM Code'].strip()
    val = row['Comorbidity Description'].strip()
    if ':' in val:
        val = val.split(': ', 1)
        val = val[1]
    if key in nta_dict:
        nta_dict[key].append(val)
    else:
        nta_dict[key] = [val]
```

```python
for key, value in nta_dict.items():
    print("'" + str(key) + "'" + ': ' + str(value) + ',')
```
